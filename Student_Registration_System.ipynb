```python
# Student Registration System

# Get student details from the user
student_name = input("Enter Student Name: ")
student_age = int(input("Enter Student Age: "))
course_name = input("Enter Course Name: ")
course_fee = float(input("Enter Course Fee: "))

# Calculate GST and Total Fee
gst = course_fee * 0.18
total_fee = course_fee + gst

# Display student details using f-string
print("\n----- Student Registration Details -----")
print(f"Student Name : {student_name}")
print(f"Student Age  : {student_age}")
print(f"Course Name  : {course_name}")
print(f"Course Fee   : ₹{course_fee:.2f}")
print(f"GST (18%)    : ₹{gst:.2f}")
print(f"Total Fee    : ₹{total_fee:.2f}")

# Display data types
print("\n----- Data Types -----")
print(f"Data type of student_name: {type(student_name)}")
print(f"Data type of student_age : {type(student_age)}")
print(f"Data type of course_name : {type(course_name)}")
print(f"Data type of course_fee  : {type(course_fee)}")
print(f"Data type of gst         : {type(gst)}")
print(f"Data type of total_fee   : {type(total_fee)}")
```

```python
# Library Fine Calculator

# Get details from the user
student_name = input("Enter Student Name: ")
book_name = input("Enter Book Name: ")
days_late = int(input("Enter Number of Days Late: "))
fine_per_day = float(input("Enter Fine Per Day: "))

# Calculate total fine
total_fine = days_late * fine_per_day

# Display details
print("\n----- Library Fine Details -----")
print(f"Student Name : {student_name}")
print(f"Book Name    : {book_name}")
print(f"Days Late    : {days_late}")
print(f"Fine Per Day : ₹{fine_per_day:.2f}")
print(f"Total Fine   : ₹{total_fine:.2f}")

# Display data types
print("\n----- Data Types -----")
print(f"Data type of student_name : {type(student_name)}")
print(f"Data type of book_name    : {type(book_name)}")
print(f"Data type of days_late    : {type(days_late)}")
print(f"Data type of fine_per_day : {type(fine_per_day)}")
print(f"Data type of total_fine   : {type(total_fine)}")
```
