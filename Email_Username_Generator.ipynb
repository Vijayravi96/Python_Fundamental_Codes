```python
# Case Study – Email Username Generator

# Taking input from the user
full_name = input("Enter Full Name: ")
domain_name = input("Enter Domain Name (e.g., gmail.com): ")

# Convert name to lowercase
full_name_lower = full_name.lower()

# Replace spaces with dots
username = full_name_lower.replace(" ", ".")

# Split the name into parts
name_parts = full_name_lower.split()

# Join the name parts using join()
joined_name = ".".join(name_parts)

# Create email address
email = f"{joined_name}@{domain_name}"

# Display output
print(f"\nFull Name: {full_name}")
print(f"Generated Email: {email}")
```
